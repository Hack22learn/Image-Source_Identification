# Source Identification using 3 Class of image

```python
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import normalize
import numpy as np
import pandas as pd
#import matplotlib.pyplot as plt
```

## Understanding Data 

```python
sony=pd.read_csv("./data/Sony.csv")
sony.head()
```

```python
nikon=pd.read_csv("./data/Nikkon.csv")
nikon.head()
```

```python
canon=pd.read_csv("./data/Canon.csv")
canon.head()
```

```python
nikon.describe()
```

```python
sony.describe()
```

```python
canon.describe()
```

# Create trainig and testing data Frame

### Add Label Coloumn

```python
sony.loc[:,'label'] = pd.Series(0, index=sony.index)
```

```python
sony.head()
```

```python
nikon.loc[:,'label'] = pd.Series(1, index=nikon.index)
nikon.head()
```

```python
canon.loc[:,'label'] = pd.Series(2, index=canon.index)
canon.head()
```

## Join ALL data set & Normalise

```python
data=pd.concat([sony,nikon,canon])
data[140:148]
```

```python
#select all coulomn except label
data[1:3].loc[:,data.columns != 'label']
```

```python
Norm_feature=normalize(data.loc[:,data.columns != 'label'], norm='l2', axis=1, copy=True)
```

```python
Norm_feature,len(Norm_feature)
```

# Create train and test

```python
#concatenate numpy array

#training Set
train_x=np.concatenate((Norm_feature[15:144], Norm_feature[160:286],Norm_feature[301:]), axis=0)
train_y=np.concatenate((np.array(data[15:144].loc[:,"label"]),np.array(data[160:286].loc[:,"label"]),
                       np.array(data[301:].loc[:,"label"])),axis=0)

#Testing Set
test_x=np.concatenate((Norm_feature[0:15], Norm_feature[144:161],Norm_feature[286:301]), axis=0)
test_y=np.concatenate((np.array(data[0:15].loc[:,"label"]),np.array(data[144:161].loc[:,"label"]),
                      np.array(data[286:301].loc[:,"label"])),axis=0)
```

```python
#concatenate numpy array

#training Set
train_x=np.concatenate((Norm_feature[15:144], Norm_feature[160:286],Norm_feature[301:]), axis=0)
train_y=np.concatenate((np.array(data[15:144].loc[:,"label"]),np.array(data[160:286].loc[:,"label"]),
                       np.array(data[301:].loc[:,"label"])),axis=0)

#Testing Set
test_x=np.concatenate((Norm_feature[0:20], Norm_feature[144:165],Norm_feature[286:306]), axis=0)
test_y=np.concatenate((np.array(data[0:20].loc[:,"label"]),np.array(data[144:165].loc[:,"label"]),
                      np.array(data[286:306].loc[:,"label"])),axis=0)
```

```python
print "taining data :",len(train_x),len(train_y)
print "testing_data :",len(test_x),len(test_y)
```

# SVM Classifier

```python
# clsf = classifier
clsf=SVC(kernel='rbf',gamma=60,C=4) #SVM Classier
svm={} #store accuracy data for plotting
```

```python
#training
clsf.fit(train_x, train_y)
```

### Testing accuracy on training data

```python
#Now testing for trainig data
#Now predict values for given classifier
prediction = clsf.predict(train_x)
svm["trainig_set"]=accuracy_score(prediction,train_y)*100
print 'Accuracy Check ',svm["trainig_set"],'%'
```

### Testing accuracy on Testing data

```python
prediction = clsf.predict(test_x)
svm["test_set"]=accuracy_score(prediction,test_y)*100
print 'Accuracy Check ',svm["test_set"],'%'
```

```python
svm
```

# Decision Tree Classifier

```python
decision={}
clf=DecisionTreeClassifier(max_depth=15,min_samples_split=3)
clf
```

```python
#train
clf = clf.fit(train_x, train_y)
```

### Testing accuracy on training data

```python
#Now testing for trainig data
#Now predict values for given classifier
prediction = clf.predict(train_x)
decision["trainig_set"]=accuracy_score(prediction,train_y)*100
print 'Accuracy Check ',decision["trainig_set"],'%'
```

### Testing accuracy on Testing data

```python
prediction = clf.predict(test_x)
decision["test_set"]=accuracy_score(prediction,test_y)*100
print 'Accuracy Check ',decision["test_set"],'%'
```

```python
decision
```

### Testing accuracy on Testing data

```python
prediction = clf.predict(test_x)
neural["test_set"]=accuracy_score(prediction,test_y)*100
print 'Accuracy Check ',neural["test_set"],'%'
```
