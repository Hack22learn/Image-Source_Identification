# Classification image using block based extracted features - 2class

```python
import numpy as np
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import normalize
import pandas as pd
```

## Import data

#### Nikon

```python
nikon=pd.read_csv("Final_dataset/Nikkon.csv")
nikon.head()
```

```python
# describe datset
nikon.describe()
```

#### Canon

```python
canon=pd.read_csv("Final_dataset/canon.csv")
canon.head()
```

```python
#describe data
canon.describe()
```

## Adding Class label

<ul>
<li>Canon : 0</li>
<li>Nikon : 1</li>

</ul>

```python
nikon.loc[:,'label'] = pd.Series(1, index=nikon.index)
nikon.head()
```

```python
canon.loc[:,'label'] = pd.Series(0, index=canon.index)
canon.head()
```

## Join Both data set & Normalise

```python
data=pd.concat([canon,nikon])
data[595:605]
```

```python
#select all coulomn except label #testing
data[1:3].loc[:,data.columns != 'label']
```

### Normalise 

```python
Norm_feature=normalize(data.loc[:,data.columns != 'label'], norm='l2', axis=1, copy=True)
```

```python
Norm_feature
```

```python
print(len(Norm_feature))
```

# Create train and test

```python
#concatenate numpy array

#training Set
train_x=Norm_feature[80:1100]
train_y=np.array(data[80:1100].loc[:,"label"])

#Testing Set
test_x=np.concatenate((Norm_feature[0:100], Norm_feature[1100:]), axis=0)
test_y=np.concatenate((np.array(data[0:100].loc[:,"label"]),np.array(data[1100:].loc[:,"label"])),axis=0)
```

# Classification

## SVM Classifier 

```python
# clsf = classifier
clsf=SVC(kernel='rbf',gamma=60,C=5) #SVM Classier
svm={} #store accuracy data for plotting
```

```python
#training
clsf.fit(train_x, train_y)
```

### Testing accuracy on training data

```python
#Now testing for trainig data
#Now predict values for given classifier
prediction = clsf.predict(train_x)
svm["trainig_set"]=accuracy_score(prediction,train_y)*100
print 'Accuracy Check ',svm["trainig_set"],'%'
```

### Testing accuracy on Testing data

```python
prediction = clsf.predict(test_x)
svm["test_set"]=accuracy_score(prediction,test_y)*100
print 'Accuracy Check ',svm["test_set"],'%'
```

```python
svm
```

## Decision Tree Classifier

```python
decision={}
clf=DecisionTreeClassifier(max_depth=50,min_samples_split=12)
clf
```

### Testing accuracy on Testing data
