# Classification image using block based extracted features - 3class

```python
import numpy as np
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import normalize
import pandas as pd
```

## Import data

#### Nikon

```python
nikon=pd.read_csv("Final_dataset/Nikkon.csv")
nikon.head()
```

#### Canon

```python
canon=pd.read_csv("Final_dataset/canon.csv")
canon.head()
```

```python
#describe data
canon.describe()
```

```python
sony=pd.read_csv("Final_dataset/Sony.csv")
canon.head()
```

```python
#describe data
sony.describe()
```

## Adding Class label

<ul>
<li>Canon : 0</li>
<li>Nikon : 1</li> 
<li>Sony : 2</li>

</ul>

```python
nikon.loc[:,'label'] = pd.Series(1, index=nikon.index)
nikon.head()
```

```python
canon.loc[:,'label'] = pd.Series(0, index=canon.index)
canon.head()
```

```python
sony.loc[:,'label'] = pd.Series(2, index=sony.index)
sony.head()
```

## Join ALL data set & Normalise

```python
data=pd.concat([canon,nikon,sony])
data[595:605]
```

```python
#select all coulomn except label #testing
data[1:3].loc[:,data.columns != 'label']
```

### Normalise 

```python
Norm_feature=normalize(data.loc[:,data.columns != 'label'], norm='l2', axis=1, copy=True)
```

```python
Norm_feature
```

```python
print(len(Norm_feature))
```

# Create train and test

```python
#concatenate numpy array

#training Set
train_x=np.concatenate((Norm_feature[60:1140],Norm_feature[1240:]),axis=0)
train_y=np.concatenate((np.array(data[60:1140].loc[:,"label"]),np.array(data[1240:].loc[:,"label"])),axis=0)

# Testing Set
test_x=np.concatenate((Norm_feature[0:100], Norm_feature[1100:1300]), axis=0)
test_y=np.concatenate((np.array(data[0:100].loc[:,"label"]),np.array(data[1100:1300].loc[:,"label"])),axis=0)
```

```python
len(test_x)
```

# Classification

## SVM Classifier 

```python
# clsf = classifier
clsf=SVC(kernel='rbf',gamma=15,C=10) #SVM Classier
svm={} #store accuracy data for plotting
```

```python
#training
clsf.fit(train_x, train_y)
```

### Testing accuracy on training data

```python
#Now testing for trainig data
#Now predict values for given classifier
prediction = clsf.predict(train_x)
svm["trainig_set"]=accuracy_score(prediction,train_y)*100
print 'Accuracy Check ',svm["trainig_set"],'%'
```

### Testing accuracy on Testing data

```python
prediction = clsf.predict(test_x)
svm["test_set"]=accuracy_score(prediction,test_y)*100
print 'Accuracy Check ',svm["test_set"],'%'
```

```python
svm
```

## Decision Tree Classifier

```python
decision={}
clf=DecisionTreeClassifier(max_depth=50,min_samples_split=12)
clf
```

```python
#train
clf = clf.fit(train_x, train_y)
```

### Testing accuracy on training data

```python
#Now testing for trainig data
#Now predict values for given classifier
prediction = clf.predict(train_x)
decision["trainig_set"]=accuracy_score(prediction,train_y)*100
print 'Accuracy Check ',decision["trainig_set"],'%'
```

### Testing accuracy on Testing data

```python
prediction = clf.predict(test_x)
decision["test_set"]=accuracy_score(prediction,test_y)*100
print 'Accuracy Check ',decision["test_set"],'%'
```
